```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\user\Downloads\creditcard.csv')
df
```

```python
df['Class'].value_counts()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
amount=df['Amount'].values
df['Amount']=sc.fit_transform(amount.reshape(-1,1))
```

```python
df
```

```python
df.drop('Time',axis=1,inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model1.fit(x_train,y_train)
```

```python
y_pred=model1.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
#The Accuracy of Logistic regression is 99.72%
```

```python
from sklearn.neighbors import KNeighborsClassifier
n=7
KNN=KNeighborsClassifier(n_neighbors=n)
KNN.fit(x_train,y_train)
```

```python
y_predK=KNN.predict(x_test)
y_predK
```

```python
accuracy_score(y_test,y_predK)
```

```python
#The Accuracy of Kmeans is 99.72%
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Dt=DecisionTreeClassifier(max_depth = 4, criterion = 'entropy')
```

```python
Dt.fit(x_train,y_train)
y_predD=Dt.predict(x_test)
y_predD
```

```python
accuracy_score(y_test,y_predD)
```
